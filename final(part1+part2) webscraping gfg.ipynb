## Merge them

```python


# Load the two DataFrames from the CSV files
df = pd.read_csv("gfgy.csv")
df1 = pd.read_csv("gfgyt.csv")

# Merge the two DataFrames based on the 'title' and 'video_link' columns
merged_df = pd.merge(df, df1, on=['title'])
```

```python
merged_df
```

```python
columns_to_drop = ["date_time_x", "views", "video_link", "like"]
merged_df = merged_df.drop(columns=columns_to_drop)
```

## need to change datatypes

```python
data_types = df.dtypes

print(data_types)
```

```python
merged_df['date_time_y'] = pd.to_datetime(merged_df['date_time_y'])
```

```python
merged_df['view'] = merged_df['view'].str.replace(',', '').str.extract(r'(\d+)').astype(int)
```

```python
data_types = merged_df.dtypes
print(data_types)
```

```python
def convert_to_hh_mm_ss(duration_str):
    if ':' in duration_str:
        parts = duration_str.split(':')
        if len(parts) == 2:
            return '00:' + duration_str
    return duration_str

# Convert 'mm:ss' to '00:mm:ss'
merged_df['length'] = merged_df['length'].apply(convert_to_hh_mm_ss)

# Convert the 'length' column to timedelta format
try:
    merged_df['length'] = pd.to_timedelta(merged_df['length'])
    print(merged_df.head(10))
except ValueError as e:
    print("Error:", e)
```

```python
merged_df
```

```python
def format_timedelta(timedelta_val):
    hours, remainder = divmod(timedelta_val.seconds, 3600)
    minutes, seconds = divmod(remainder, 60)
    return f"{hours:02d}:{minutes:02d}:{seconds:02d}"

# Apply the function to the 'length' column
merged_df['length'] = merged_df['length'].apply(format_timedelta)

print(merged_df.head(10))
```

## 1. Number of videos in the past 6 months from the start date. Must mention the dates in the solution.

```python

today = pd.to_datetime('2023-08-07')  # Replace '2023-08-07' with the current date if running in the future
six_months_ago = today - pd.DateOffset(months=6)
videos_within_six_months = merged_df[merged_df['date_time_y'] >= six_months_ago]
num_videos_within_six_months = len(videos_within_six_months)
print("Number of videos in the past 6 months:", num_videos_within_six_months)
print("Start date of the 6-month period:", six_months_ago.date())
```

## 2. Create a pandas data frame with columns name as videos title, views, Length of videos, and videos link

```python
new_df = merged_df[['title', 'view', 'length', 'link']].copy()
new_df
```

## 3. Name the most viewed topics in the past 6 months

```python
most_viewed_topics = new_df.sort_values(by='view', ascending=False).head(5)
print("\nMost viewed topics in the past 6 months:")
most_viewed_topics
```

## 4. Name the topics with the highest video length.

```python
new_df['length_seconds'] = pd.to_timedelta(new_df['length']).dt.total_seconds()
topics_with_highest_length = new_df.sort_values(by='length_seconds', ascending=False).head(5)
print("\nTopics with the highest video length:")
print(topics_with_highest_length[['title', 'length']])
```

## 5 Make a comparison between the number of views and video length using a Graph

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.scatter(new_df['length_seconds'], new_df['view'], alpha=0.5)
plt.xlabel('Video Length (seconds)')
plt.ylabel('Number of Views')
plt.title('Number of Views vs. Video Length')

plt.subplot(1, 2, 2)
plt.hist(new_df['view'], bins=30)
plt.xlabel('Number of Views')
plt.ylabel('Frequency')
plt.title('Distribution of Number of Views')

plt.tight_layout()
plt.show()
```
