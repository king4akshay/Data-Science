```python
import time
import numpy as np
import pandas as pd
from tqdm import tqdm
from bs4 import BeautifulSoup
from selenium import webdriver
```

```python
browser = webdriver.Chrome()
browser.get('https://www.youtube.com/@GeeksforGeeksVideos/videos')
```

```python
soup = BeautifulSoup(browser.page_source, "html.parser")
print(soup.text.replace("\n\n",""))
```

```python
data = []
for sp in tqdm(soup.find_all("ytd-rich-item-renderer")):
    title=sp.find("a",class_="yt-simple-endpoint focus-on-expand style-scope ytd-rich-grid-media").text
    try:
        views=sp.find_all("span",class_="ytd-video-meta-block")[0].text
    except:
        views  = np.nan
    try:
        length =sp.find("span", class_="style-scope ytd-thumbnail-overlay-time-status-renderer").text
    except:
        length  = np.nan
    try:   
        video_link = sp.find("a", class_="yt-simple-endpoint focus-on-expand style-scope ytd-rich-grid-media").get("href")
    except:
        video_link  = np.nan
    try:
        date_time=sp.find_all("span",class_="ytd-video-meta-block")[1].text
    except:
        date_time  = np.nan
    data.append([title, views,length, video_link, date_time])
   
```

```python
data
```

```python
df = pd.DataFrame(data, columns = ['title','views','length','video_link','date_time'])
```
